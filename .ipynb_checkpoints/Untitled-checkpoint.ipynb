```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import rawpy

mpl.rcParams["axes.grid"] = False

%matplotlib inline
```

```python
raw = rawpy.imread("./sample.ARW")
```

```python
img_preview = raw.postprocess(use_camera_wb=True)
```

```python
plt.imshow(img_preview)
```

```python
def black_correction(src, black_array):
    dst = src.copy()
    dst[0::2,0::2] -= black_array[0]
    dst[0::2,1::2] -= black_array[1]
    dst[1::2,0::2] -= black_array[2]
    dst[1::2,1::2] -= black_array[3]
    dst[dst < 0] = 0
    return dst
```

```python
def demosic(src):
    h, w = src.shape
    dst = np.zeros((h//2) * (w//2) * 3).reshape(h//2, w//2, 3)
    dst[:,:,0] = src[0::2,0::2] 
    dst[:,:,1] = (src[0::2,1::2] + src[1::2,0::2]) / 2
    dst[:,:,2] = src[1::2,1::2] 
    return dst
```

```python
def white_balance(src, wb):
    dst = src.copy()
    dst[:,:,0] = src[:,:,0] * (wb[0] / wb[1])
    dst[:,:,1] = src[:,:,1] * (wb[1] / wb[1])
    dst[:,:,2] = src[:,:,2] * (wb[2] / wb[1])    
    return dst
```

```python
def color_matrix(src, color_matrix):
    dst = src.copy()
    dst[:,:,0] = src[:,:,0] * color_matrix[0,0] + src[:,:,1] * color_matrix[0,1] + src[:,:,2] * color_matrix[0,2]
    dst[:,:,1] = src[:,:,0] * color_matrix[1,0] + src[:,:,1] * color_matrix[1,1] + src[:,:,2] * color_matrix[1,2]
    dst[:,:,2] = src[:,:,0] * color_matrix[2,0] + src[:,:,1] * color_matrix[2,1] + src[:,:,2] * color_matrix[2,2]    
    return dst
```

```python
def standarization(src):
    dst = src.copy()
    dst[dst<0] = 0
    dst /= dst.max()
    return dst
```

```python
def gamma_correction(src, r):
    dst = src.copy()
    dst = standarization(dst)
    dst = dst ** (1 / r)
    return dst
```

```python
h, w = raw.sizes.raw_height, raw.sizes.raw_width
img_raw = raw.raw_image.copy().astype(np.float)
```

```python
mtx = np.array([[1141, -205, 88], [-52, 1229, -154], [70, -225, 1179]]) / 1024
```

```python
img_raw = black_correction(img_raw, raw.black_level_per_channel)
img = demosic(img_raw)
img = white_balance(img, raw.camera_whitebalance)
img = color_matrix(img, mtx)
img = gamma_correction(img, r=2.2)
img = standarization(img)
```
